```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import seaborn as sns
# import pandas as pd

df = sns.load_dataset("tips")
df.head()
```

```python
# plt.plot("total_bill")
# sns.lineplot(df["total_bill"])
sns.scatterplot(df["total_bill"])
sns.scatterplot(x = "total_bill",y="tip",data=df)
# plt.title("Scatterplot: Total Bill vs Tip")
# plt.show()
```

```python
sns.histplot(df["total_bill"],color= "red", edgecolor = "blue")
```

```python
## outliers handle
q1 = df["total_bill"].quantile(0.25)
q2 = df["total_bill"].quantile(0.75)
iqr = q2-q1
lower_limit = q1-1.5*iqr
upper_limit = q2+1.5*iqr
print(lower_limit, upper_limit)
# print("Lower Limit:", lower_limit)
# print("Upper Limit:", upper_limit)
# df[df["total_bill"] < lower_limit]
df[df["total_bill"] > upper_limit]
df = df[(df["total_bill"] >= lower_limit) & (df["total_bill"] <= upper_limit)]
sns.boxplot(df["total_bill"])
```
