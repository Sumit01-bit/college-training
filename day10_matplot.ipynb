```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
arr = np.random.randint(20000,40000,100)
arr
```

```python
## lineplot
arr.sort()
plt.plot(arr)
# plt.show()
```

```python
x = [2,5, 5, 2, 2, 5]
y = [1,2, 15, 15, 25, 25]

arr.sort()
plt.plot(x,y,marker = "s",ms = 10,mfc = "red",mec = "green",color = "blue" , lw = 3,ls = ":",label = "lineplot")
plt.title("tittle")
plt.xlabel("x-axies")
plt.ylabel("y-axies")
plt.legend()
plt.show()
```

```python
#scatterplot
plt.scatter(x,y,marker="D")
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
colours = ["red","black","blue","yellow"]
hatch_pattern = ["//","\\\\","xx","oo"]
plt.bar(x,y,color = colours,hatch = hatch_pattern,label = x,width=0.3,edgecolor = "black",alpha = 0.7,linewidth = 3)
# plt.barh(x,y)
plt.legend()
plt.show()
```

```python
##histogram -- distribution of data in interva
##piechar -- catergorial 
arr = np.random.randint(20000,40000,100)
plt.hist(arr,bins=20)
plt.show()
```

```python

x = ["a", "b", "c", "d"]
y = [10, 20, 30, 40]
explode = (0.4, 0, 0, 0)
plt.pie(y, labels=x, autopct="%1.2f%%", shadow=True, explode=explode)
plt.savefig("save.png")
plt.show()
```
